```python
import librosa
from dotenv import load_dotenv
import assemblyai as aai
import os

# Load the .env file
load_dotenv()

aai.settings.api_key = os.getenv("API_Key")
```

```python
#path to audio file
audio_file = "data/afjiv.wav"

config = aai.TranscriptionConfig(
    speech_model=aai.SpeechModel.universal,
    speaker_labels=True 
)

transcript = aai.Transcriber(config=config).transcribe(audio_file)

if transcript.status == "error":
    raise RuntimeError(f"Transcription failed: {transcript.error}")

# Print with speaker labels
# print("\n" + "="*80)
# print(f"{'Start':<10} {'End':<10} {'Speaker':<12} {'Text'}")
# print("="*80)

# for utterance in transcript.utterances:
#     start_time = utterance.start / 1000  
#     end_time = utterance.end / 1000
#     speaker = utterance.speaker
#     text = utterance.text
    
#     print(f"{start_time:<10.2f} {end_time:<10.2f} Speaker {speaker:<5} {text}")

# print("="*80)

basename = os.path.splitext(os.path.basename(audio_file))[0]
out_path = f"{basename}_transcript.txt"

with open(out_path, "w", encoding="utf-8") as f:
    for u in transcript.utterances:
        start = u.start / 1000
        end = u.end / 1000
        speaker = u.speaker
        text = u.text.replace("\n", " ")
        f.write(f"[ Start Time:{start} End Time:{end} ]\nSpeaker {speaker}: {text}\n")


print(f"Wrote transcript to {out_path}")
```
